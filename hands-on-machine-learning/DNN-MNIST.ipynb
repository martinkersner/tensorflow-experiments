```python
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
X = mnist.train.images
y = np.argmax(mnist.train.labels, axis=1)

X_test = mnist.test.images
y_test = np.argmax(mnist.test.labels, axis=1)

scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X)
X_test_norm = scaler.transform(X_test)

feature_columns = tf.contrib.learn.infer_real_valued_columns_from_input(X_train_norm)

# 2 hidden layers with 300 and 100 units
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300, 100], n_classes=10,
                                         feature_columns=feature_columns)
```

```python
dnn_clf.fit(x=X_train_norm, y=y.reshape((-1, 1)), batch_size=50, steps=40000)
```

```python
y_pred = list(dnn_clf.predict(X_test_norm))
```

```python
accuracy_score(list(y_test), y_pred)
```
